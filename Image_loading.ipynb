```python
def validate_image(filepath: str) -> bool:
    """
    Check if a file is a valid image.
    
    Args:
        filepath: Path to image file
    
    Returns:
        True if valid image, False otherwise
    """
    try:
        with Image.open(filepath) as img:
            img.verify()
        return True
    except:
        return False
```
